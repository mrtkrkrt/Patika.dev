**Prepare train, val, test datas**

```python
X_heat = df.drop(columns=["Y1", "Y2"])
Y_heat = df["Y1"]
```

```python
X_cool = df.drop(columns=["Y1", "Y2"])
Y_cool = df["Y2"]
```

**Split Data train, test, validation**

```python
from sklearn.model_selection import train_test_split
x_heat_train, x_heat_val, y_heat_train, y_heat_val = train_test_split(X_heat, Y_heat, test_size = 0.15, random_state = 42)
x_heat_val, x_heat_test, y_heat_val, y_heat_test = train_test_split(x_heat_val, y_heat_val, test_size = 0.5, random_state = 42)
```

```python
x_cool_train, x_cool_val, y_cool_train, y_cool_val = train_test_split(X_cool, Y_cool, test_size = 0.15, random_state = 42)
x_cool_val, x_cool_test, y_cool_val, y_cool_test = train_test_split(x_cool_val, y_cool_val, test_size = 0.5, random_state = 42)
```

```python
from lazypredict.Supervised import LazyRegressor
lzr = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
models_heat, preditions_heat = lzr.fit(x_heat_train, x_heat_val, y_heat_train, y_heat_val)
```
